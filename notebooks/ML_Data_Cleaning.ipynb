```python
# ============================================
# BANK GROUP RATIOS - DATA CLEANING
# Step 1: Explore the file structure
# ============================================

import pandas as pd
import numpy as np

# Load the file from raw_data folder
file_path = r"D:\Bank_EWS_Project\raw_data\bank_group_ratios.xlsx"

print("="*60)
print("STEP 1: READING RAW FILE")
print("="*60)

# Read without header to see all rows
df_raw = pd.read_excel(file_path, header=None)

print(f"\n📊 Shape: {df_raw.shape}")
print(f"   Rows: {df_raw.shape[0]}")
print(f"   Columns: {df_raw.shape[1]}")

print("\n📋 FIRST 10 ROWS:")
print(df_raw.head(10))

print("\n" + "="*60)
print("Look at the output above and answer:")
print("1. How many rows and columns do you see?")
print("2. At which row do the actual column headers appear?")
print("3. Do you see 'Return on assets' or 'Return on equity' anywhere?")
print("="*60)
```

```python
# ============================================
# STEP 2: Load with correct headers
# ============================================

# Read the file with headers at row 6 (index 6)
df = pd.read_excel(file_path, header=6)

print("="*60)
print("STEP 2: DATA WITH CORRECT HEADERS")
print("="*60)

print(f"\n📊 Shape: {df.shape}")
print(f"   Rows: {df.shape[0]}")
print(f"   Columns: {df.shape[1]}")

print("\n📋 COLUMN NAMES:")
for i, col in enumerate(df.columns):
    print(f"   {i}: {col}")

print("\n📋 FIRST 5 ROWS:")
print(df.head())

print("\n📋 LAST 5 ROWS:")
print(df.tail())

print("\n" + "="*60)
```

```python
# ============================================
# STEP 5: Find exact ROA and ROE rows
# ============================================

print("="*60)
print("STEP 5: LOCATING ROA AND ROE EXACTLY")
print("="*60)

# Look for rows containing 'Return on assets' (ROA)
roa_rows = df[df['Ratios'].str.contains('Return on assets', case=False, na=False)]
print(f"\n🔍 Return on assets (ROA) found: {len(roa_rows)} rows")
if len(roa_rows) > 0:
    print(roa_rows[['Year', 'Ratios', 'PUBLIC SECTOR BANKS', 'PRIVATE SECTOR BANKS', 
                    'FOREIGN BANKS', 'ALL SCHEDULED COMMERICIAL BANKS']])
else:
    print("   Not found - trying variations...")
    
    # Try variations
    roa_rows = df[df['Ratios'].str.contains('Return on assets', case=False, na=False, regex=False)]
    if len(roa_rows) == 0:
        roa_rows = df[df['Ratios'].str.contains('return on assets', case=False, na=False, regex=False)]

# Look for rows containing 'Return on equity' (ROE)
roe_rows = df[df['Ratios'].str.contains('Return on equity', case=False, na=False)]
print(f"\n🔍 Return on equity (ROE) found: {len(roe_rows)} rows")
if len(roe_rows) > 0:
    print(roe_rows[['Year', 'Ratios', 'PUBLIC SECTOR BANKS', 'PRIVATE SECTOR BANKS', 
                    'FOREIGN BANKS', 'ALL SCHEDULED COMMERICIAL BANKS']])
else:
    print("   Not found")
    
# Let's check all rows that might contain ROA/ROE
print("\n🔍 Searching all ratio names:")
for idx, row in df.iterrows():
    ratio = str(row['Ratios'])
    if 'return' in ratio.lower() or 'assets' in ratio.lower() or 'equity' in ratio.lower():
        if 'Return on' in ratio or 'return on' in ratio.lower():
            print(f"Row {idx}: {ratio}")
```

```python
# ============================================
# STEP 8a: Check where years are stored
# ============================================

print("="*60)
print("STEP 8a: FINDING THE YEARS")
print("="*60)

# Look at the original dataframe again to see Year column
print("\n📋 FIRST 10 ROWS OF ORIGINAL DATA (Year column):")
for i in range(10):
    print(f"Row {i}: Year = '{df.iloc[i]['Year']}'")

print("\n📋 ROWS 20-25 (where we found ROA):")
for i in range(20, 26):
    print(f"Row {i}: Year = '{df.iloc[i]['Year']}', Ratio = '{df.iloc[i]['Ratios']}'")
```

```python
# ============================================
# STEP 8b: Extract years from the data
# ============================================

print("="*60)
print("STEP 8b: EXTRACTING YEARS")
print("="*60)

# Create a list to store years
years = []
current_year = None

# Go through each row and capture years
for idx, row in df.iterrows():
    year_val = row['Year']
    
    # If year is not NaN and not empty, it's a new year
    if pd.notna(year_val) and str(year_val).strip() != '' and str(year_val).strip() != 'nan':
        current_year = str(year_val).strip()
        print(f"Row {idx}: Found year '{current_year}'")
    
    years.append(current_year)

# Add the year column to our dataframe
df['Extracted_Year'] = years

print(f"\n✅ Added 'Extracted_Year' column to dataframe") 
```

```python
# ============================================
# STEP 8c: Extract ROA with years
# ============================================

print("="*60)
print("STEP 8c: EXTRACTING ROA WITH YEARS")
print("="*60)

# Find ROA rows (ratio 21)
roa_rows = []
for idx, row in df.iterrows():
    ratio = str(row['Ratios'])
    if '21.  Return on assets' in ratio:
        roa_rows.append({
            'Year': row['Extracted_Year'],
            'ROA_Public': row['PUBLIC SECTOR BANKS'],
            'ROA_Private': row['PRIVATE SECTOR BANKS'],
            'ROA_Foreign': row['FOREIGN BANKS'],
            'ROA_All': row['ALL SCHEDULED COMMERICIAL BANKS']
        })

# Convert to dataframe
roa_df = pd.DataFrame(roa_rows)
print(f"\n📊 Found {len(roa_df)} ROA rows")
print("\n📋 ROA data with years:")
print(roa_df)
```

```python
# ============================================
# STEP 8d: Extract ROE with years
# ============================================

print("="*60)
print("STEP 8d: EXTRACTING ROE WITH YEARS")
print("="*60)

# Find ROE rows (ratio 22)
roe_rows = []
for idx, row in df.iterrows():
    ratio = str(row['Ratios'])
    if '22.  Return on equity' in ratio:
        roe_rows.append({
            'Year': row['Extracted_Year'],
            'ROE_Public': row['PUBLIC SECTOR BANKS'],
            'ROE_Private': row['PRIVATE SECTOR BANKS'],
            'ROE_Foreign': row['FOREIGN BANKS'],
            'ROE_All': row['ALL SCHEDULED COMMERICIAL BANKS']
        })

# Convert to dataframe
roe_df = pd.DataFrame(roe_rows)
print(f"\n📊 Found {len(roe_df)} ROE rows")
print("\n📋 ROE data with years:")
print(roe_df)
```

```python
# ============================================
# STEP 8e: Merge ROA and ROE
# ============================================

print("="*60)
print("STEP 8e: MERGING ROA AND ROE")
print("="*60)

# Merge on Year
bank_ratios = pd.merge(roa_df, roe_df, on='Year', how='outer')

print(f"\n📊 Final shape: {bank_ratios.shape}")
print(f"   Rows: {bank_ratios.shape[0]}")
print(f"   Columns: {bank_ratios.shape[1]}")

print("\n📋 FINAL DATA (with years):")
print(bank_ratios)

print("\n📊 Check if we have all 21 years:")
print(bank_ratios['Year'].tolist())  
```

```python
# Quick verification
print("✅ Data saved successfully!")
print(f"Total years: {len(bank_ratios)}")
print(f"Year range: {bank_ratios['Year'].min()} to {bank_ratios['Year'].max()}")
```
